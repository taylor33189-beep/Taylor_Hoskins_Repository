```python
import pandas as pd
import statsmodels.api as sm
from io import StringIO

file_path = 'attendance-2016.csv'
data = pd.read_csv(file_path)

# Normalize column names
data.columns = data.columns.str.lower()

columns_to_process = ['won', 'runspergame', 'homeavgatt']
df = data[columns_to_process].apply(pd.to_numeric, errors='coerce').dropna()

# 1: Correlation Analysis
correlation_matrix = df.corr(method='pearson')
corr_output = correlation_matrix.to_string()

# 2: SLR: Wins on RunsPerGame
Y_slr = df['won']
X_slr = sm.add_constant(df['runspergame'])
slr_model = sm.OLS(Y_slr, X_slr).fit()
slr_summary = slr_model.summary()

# 3: MLR: HomeAvgAtt on Won and RunsPerGame
Y_mlr = df['homeavgatt']
X_mlr = sm.add_constant(df[['won', 'runspergame']])
mlr_model = sm.OLS(Y_mlr, X_mlr).fit()
mlr_summary = mlr_model.summary()

print("Correlation Matrix")
print(corr_output)

print("\nSLR: Wins on RunsPerGame Summary")
print(slr_summary.as_text())

print("\nMLR: HomeAvgAtt on Won and RunsPerGame Summary")
print(mlr_summary.as_text())
```
